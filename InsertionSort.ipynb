```python
def insertion_sort(arr):
    for i in range(1, len(arr)):
        key = arr[i]
        j = i - 1
        while j >= 0 and arr[j] > key:
            arr[j + 1] = arr[j]
            j -= 1
        arr[j + 1] = key
```

```python
arr = [9, 4, 7, 2, 1, 8, 5, 3, 6]
insertion_sort(arr)
```

```python
print("Sorted list:", arr)
```
