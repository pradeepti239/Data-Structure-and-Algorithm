##**Stack Implementation using Linked List**

```python
class Node:
  def __init__(self, item):
    self.item = item
    self.next = None
```

```python
class Stack:
  def __init__(self):
    self.top = None # Stack is empty in the beginning
    self.n = 0 # ?

  def size(self):
    return self.n

  def isempty(self):
    return self.top == None

  def push(self, value):
    new_node = Node(value)
    new_node.next = self.top
    self.top = new_node
    self.n += 1

  def peak(self):
    if self.isempty():
      return "Stack Empty"

    return self.top.item

  def __str__(self):
    curr = self.top
    result = ''
    while curr != None:
      result = result + str(curr.item) + '\n'
      curr = curr.next
    return result

  def pop(self):
    if self.isempty():
      return "Stack Empty"
    data = self.top.item
    self.top = self.top.next
    self.n -= 1

    return data

  def reverse_string(self,input_string):
    stack = []
    reversed_string = ""

    # Push each character onto the stack
    for char in input_string:
        stack.append(char)

    # Pop characters from the stack to build the reversed string
    while stack:
        reversed_string += stack.pop()

    return reversed_string

  def copy(self):
    copy = Stack()
    curr = self.top
    arr = []

    while curr != None:
      arr.append(curr.item)
      curr = curr.next

    for _ in range(len(arr)):
      copy.push(arr.pop())

    return copy
```

```python
def operation(string, operation):

    #initializing three objects to store characters from the input string,operations and temporary stack.
    string_stack = Stack()
    operation_stack = Stack()
    temp_stack = Stack()

    #popping characters from stack and returning in reverse order
    def get_string(S):
        result = ''
        while not S.isempty():
            result = S.pop() + result
        return result

    #pushing characters of string in string stack
    for char in string:
        string_stack.push(char)

    #pushing characters of operation(uuru) in the operation stack
    for char in operation:
        operation_stack.push(char)

    while not operation_stack.isempty():
        popped_operation = operation_stack.pop()
        print(f"Input: {get_string(string_stack.copy())} , Operation: {popped_operation} ,", end=' ')
        if popped_operation == 'u':
            temp_stack.push(string_stack.pop())
        elif popped_operation == 'r' and not temp_stack.isempty():
            string_stack.push(temp_stack.pop())

        print(f"Output: {get_string(string_stack.copy())}")
```
